```python
import torch
import matplotlib.pyplot as plt
import json

import seaborn as sns
sns.set_style("whitegrid")
```

```python
stats = torch.load("inference_time_stats_cuda.pth")


plt.figure(figsize=[10*len(stats), 7])
plot_id = 0
for k, configs in stats.items():
    times = []
    for config in configs.values():
        times.append(config)
    times = torch.tensor(times)

    plot_id += 1
    plt.subplot(1, len(stats), plot_id)
#     plt.plot(times.mean(dim=1))
    plt.plot(times.flatten())
    plt.title(f"K={k}")
    plt.xlabel("config_id x 100 + run_id")
    plt.ylabel("msec")

plt.tight_layout()
```

```python
cpu_stats = torch.load("inference_time_stats_cpu.pth")
gpu_stats = torch.load("inference_time_stats_cuda.pth")




plt.figure(figsize=[10*len(stats), 7])
plot_id = 0
for k in cpu_stats.keys():
    cpu_times = torch.tensor(list(cpu_stats[k].values())) * 1000
    gpu_times = torch.tensor(list(gpu_stats[k].values()))


    plot_id += 1
    plt.subplot(1, len(stats), plot_id)
    plt.plot(cpu_times.flatten(), label="cpu")
    plt.plot(gpu_times.flatten(), label="gpu")


    plt.title(f"K={k}")
    plt.xlabel("config_id x 100 + run_id")
    plt.ylabel("msec")
    plt.legend()

plt.tight_layout()
```
